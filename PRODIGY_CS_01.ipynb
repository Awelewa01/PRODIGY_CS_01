```python
#Python program that encrypts and decrypts text using the Caesar Cipher Algorithm
```

```python
def encrypt_caeser(text, shift):      #defines an encrypt function
  encrypted_text = ""     #initailizes an empty string to store the encrypted message
  for char in text:
    if char.isalpha():
      shift_amount = shift % 26     #ensures the shift is within 0-25
      ascii_offset = 65 if char.isupper() else 97   #determines ASCII starting point: A=65, a=97
      encrypted_char = chr(((ord(char) - ascii_offset + shift_amount) % 26) + ascii_offset)       #converts letter to 0-25 index, applies shift and converts back to character
      encrypted_text = encrypted_text + encrypted_char     #appends the encrypted character to the resulting string
    else:
      encrypted_text = encrypted_text + char      #keeps non-letters unchanged
  return encrypted_text
```

```python
# function to decrypt
def decrypt_caeser2(text, shift):
  decrypted_caeser = encrypt_caeser(text, 26-shift)     #decrypt by shifting in the opposite direction
  #decrypted_caesar = encrypt_caesar(text, -shift)
  return decrypted_caeser
```

```python
#User inputs message and shift value
def main():
  message = input("Enter a message to encrypt:")      #Receives message from user
  shift = int(input("Enter a shift value:"))      #Receives shift value from user, and converts to integer
  encrypted_message = encrypt_caeser(message, shift)      #encrypts the message
  print(f"Encrypted message is: {encrypted_message}")      #dsiplays the encrypted message
  decrypted_message = decrypt_caeser2(encrypted_message, shift)      #decrypts the message
  print(f"Decrypted message is: {decrypted_message}")      #displays the decrypted message

if __name__ == "main":
  main()      #runs the program when executed
```
